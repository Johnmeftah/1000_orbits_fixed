```python
# libraries

from os import name 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math 
```

```python
# creating data using the output file by Stampede

df = pd.read_csv ("/Users/john/Desktop/4.csv", sep = '\s+')
df.head()
```

```python
# using labels to plot multiple columns by only using the number of the columns

labels = ["it", "t", "dt", "torqint_2", "torqext_2", "torqint_3", "torqext_3", "xq1", "yq1", "xq3",
            "yq3", "vxq3", "vyq3", "xq2", "yq2", "vxq2", "vyq2", "vxq1", "vyq1", "dtv",
            "dtdiffus"]

# using the dropna cmd to remove rows and columns with Null/NaN values

database = df[labels].dropna(how="any") 
vals = df.values

# to debug use ::100, 0 (100 or whatever # is to check for every 100th line)
```
